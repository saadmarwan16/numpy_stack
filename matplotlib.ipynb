```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 20, 1000)
```

```python
y = np.sin(x) + 0.2 * x
```

```python
plt.plot(x, y)
```

```python
plt.plot(x, y)
plt.xlabel('input')
plt.ylabel('output')
plt.title('my show');
```
