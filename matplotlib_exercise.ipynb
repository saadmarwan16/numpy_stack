```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
generated_integers = np.random.randint(0, high=2, size=(1500, 2))
```

```python
plotting_integers = list()
plot_colors = list()

for row in generated_integers:
    if row[0] == 0 and row[1] == 0: 
        plotting_integers.append([np.random.random() * -1, np.random.random() * -1])
        plot_colors.append(0)
    elif row[0] == 0 and row[1] == 1: 
        plotting_integers.append([np.random.random(), np.random.random() * -1])
        plot_colors.append(1)
    elif row[0] == 1 and row[1] == 0: 
        plotting_integers.append([np.random.random() * -1, np.random.random()])
        plot_colors.append(1)
    else: 
        plotting_integers.append([np.random.random(), np.random.random()])
        plot_colors.append(0)
```

```python
plotting_integers = np.array(plotting_integers)
plot_colors = np.array(plot_colors)
plt.scatter(plotting_integers[:, 0], plotting_integers[:, 1], c=plot_colors)
plt.xlabel('input signal')
plt.ylabel('output signal')
plt.title('XOR distribution')
```
