```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
hx_filt = np.array([[1, 0, -1], [2, 0, -2], [1, 0, -1]])
hy_filt = np.array([[1, 2, 1], [0, 0, 0], [-1, -2, -1]])
```

```python
hy_filt
```

```python
from PIL import Image
```

```python
img = Image.open('2021-bugatti-chiron-super-sport-300.jpg')
```

```python
plt.imshow(img)
```

```python
img_arr = np.array(img)
gray = np.mean(img_arr, axis=2)
plt.imshow(gray, cmap='gray');
```

```python
from scipy.signal import convolve2d
```

```python
gx = convolve2d(gray, hx_filt)
```

```python
gy = convolve2d(gray, hy_filt)
```

```python
results = np.sqrt((gx ** 2) + (gy ** 2))
results
```

```python
plt.imshow(results, cmap='gray');
```
